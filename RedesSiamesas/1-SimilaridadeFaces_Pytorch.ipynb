### Criando o DataSet
Seu objetivo  construir a funo `generate_pairs(images_dict, num_pairs=1000)`, que deve retornar **duas listas**:

* `genuine_pairs`: pares de **imagens da mesma pessoa**  rtulo `1`
* `impostor_pairs`: pares de **imagens de pessoas diferentes**  rtulo `0`

Cada elemento dessas listas deve ter o formato:


#### Requisitos da funo

1. **Distribua o nmero de pares de forma balanceada** entre os sujeitos (para evitar que pessoas com mais imagens dominem o dataset).
   * Dica: voc pode calcular `num_pairs // len(subjects)` para estimar quantos pares por pessoa gerar.

2. **Evite duplicatas**.

   * Por exemplo, `(img1, img2)` e `(img2, img1)` devem ser considerados o mesmo par.
   * A funo `itertools.combinations(imgs, 2)` pode ajudar aqui.

3. **Gere pares genunos** (mesmo sujeito):

   * Selecione aleatoriamente duas imagens **diferentes** da mesma pessoa.

4. **Gere pares impostores** (sujeitos diferentes):

   * Escolha duas pessoas distintas e selecione uma imagem de cada uma.

5. **Mantenha o dataset balanceado**.

   * Tente gerar quantidades semelhantes de pares genunos e impostores.
   * Dica: voc pode selecionar aleatoriamente elementos da maior lista de forma que o novo tamanho seja igual o da menor lista.  

6. **Retorne duas listas:**

   ```python
   return genuine_pairs, impostor_pairs
   ```

O que queremos, ento, so duas listas como representado abaixo:




````python
def generate_pairs(images_dict, num_pairs):
    """Gera pares balanceados de forma ultra rpida com limites de segurana."""
    genuine_set = set()
    impostor_set = set()
    subjects = list(images_dict.keys())
    
    # Pr-filtra sujeitos com pelo menos 2 imagens para evitar erros no sample
    subjects_with_min_2 = [s for s in subjects if len(images_dict[s]) >= 2]
    
    if not subjects_with_min_2:
        return [], []

    target_per_type = num_pairs // 2
    
    # --- Gerao de Pares Genunos ---
    # Limite terico para no entrar em loop infinito
    max_gen = sum(len(images_dict[s]) * (len(images_dict[s]) - 1) // 2 for s in subjects_with_min_2)
    actual_gen_target = min(target_per_type, max_gen)
    
    while len(genuine_set) < actual_gen_target:
        sub = random.choice(subjects_with_min_2)
        img1, img2 = random.sample(images_dict[sub], 2)
        genuine_set.add(tuple(sorted((img1, img2))))

    # --- Gerao de Pares Impostores ---
    # Limite terico aproximado para evitar loops
    total_imgs = sum(len(v) for v in images_dict.values())
    max_imp = (total_imgs**2) // 2 
    actual_imp_target = min(target_per_type, max_imp)

    while len(impostor_set) < actual_imp_target:
        sub1, sub2 = random.sample(subjects, 2)
        img1 = random.choice(images_dict[sub1])
        img2 = random.choice(images_dict[sub2])
        impostor_set.add(tuple(sorted((img1, img2))))

    # Retorna como listas de tuplas (img1, img2, label)
    # Label 1 para genuno, 0 para impostor
    return [(p[0], p[1], 1) for p in genuine_set], [(p[0], p[1], 0) for p in impostor_set]

def save_pairs_to_csv(pairs, csv_filename):
    # Criamos o DataFrame de uma vez s para ser rpido
    df = pd.DataFrame(pairs, columns=['image1', 'image2', 'label'])
    df.to_csv(csv_filename, index=False)
    print(f" {csv_filename} salvo com {len(df)} amostras.")

def read_image_paths(root_path):
    """L as subpastas e mapeia os arquivos PGM."""
    image_map = {}
    for item in os.listdir(root_path):
        subpath = os.path.join(root_path, item)
        if os.path.isdir(subpath):
            files = [os.path.join(subpath, f).replace("\\", "/") 
                     for f in os.listdir(subpath) if f.endswith('.pgm')]
            if files:
                image_map[item] = files
    return image_map



root_path = "C:/Users/Gabriel/.cache/kagglehub/datasets/kasikrit/att-database-of-faces/versions/2/"
images_dict = read_image_paths(root_path)
all_subjects = list(images_dict.keys())

# Embaralha e divide sujeitos
np.random.shuffle(all_subjects)
n_total = len(all_subjects)
n_train, n_val = int(0.7 * n_total), int(0.15 * n_total)

train_subs = all_subjects[:n_train]
val_subs = all_subjects[n_train : n_train + n_val]
test_subs = all_subjects[n_train + n_val :]

# Configurao de gerao
splits = {
    "train": {"subs": train_subs, "n": 10000},
    "validation": {"subs": val_subs, "n": 2500},
    "test": {"subs": test_subs, "n": 1000}
}

for name, config in splits.items():
    # Cria dicionrio filtrado para o split atual
    current_dict = {s: images_dict[s] for s in config["subs"]}
    
    gen, imp = generate_pairs(current_dict, config["n"])
    save_pairs_to_csv(gen + imp, f"siamese_{name}_pairs.csv")
````

````python
class SiameseDataset(Dataset):
    def __init__(self, csv_file, device, transform=None):
        self.data = pd.read_csv(csv_file)
        self.device = device
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img1_path = self.data.iloc[idx, 0]
        img2_path = self.data.iloc[idx, 1]
        label = self.data.iloc[idx, 2]

        # Ler as imagens usando torchvision
        img1 = Image.open(img1_path).convert("L")  # L for grayscale
        img2 = Image.open(img2_path).convert("L") # L a imagem como um tensor

        if self.transform:
            img1 = self.transform(img1)
            img2 = self.transform(img2)

        # Enviar para o dispositivo (CPU ou GPU)
        img1 = img1.to(self.device)
        img2 = img2.to(self.device)
        label = torch.tensor(label).to(self.device)

        return img1, img2, label
````

````python
transform = transforms.Compose([
    transforms.Resize((100, 100)),  # Ajuste para o tamanho desejado
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # Normalizao para imagens em escala de cinza
])


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 256

train_dataset = SiameseDataset('siamese_train_pairs.csv', device, transform=transform)
validation_dataset = SiameseDataset('siamese_validation_pairs.csv', device, transform=transform)
test_dataset = SiameseDataset('siamese_test_pairs.csv', device, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=True)
````

## Criando Siamese Network

Agora que j temos nossos pares de imagens (genunos e impostores), vamos construir a rede siamesa que aprender a distinguir entre pares semelhantes e diferentes.

Uma Siamese Network  um tipo especial de arquitetura que recebe duas entradas e aprende a medir a distncia (ou dissimilaridade) entre elas.
O conceito central  que imagens da mesma classe (mesma pessoa) devem gerar representaes vetoriais prximas, enquanto imagens de classes diferentes devem gerar representaes distantes no espao de embeddings.

Nossa rede siamesa ser composta por duas partes principais:

1. Rede de Extrao de Caractersticas (BaseNetwork):
    - Ambas as imagens passam por essa mesma CNN, que gera o vetor de embedding de cada imagem.
    - Sugesto: crie um espao latente de dimenso 16, ser suficiente nesse caso.

  
2. Clculo da Distncia (ou Similaridade):
    - Aps obter os embeddings output1 e output2, calculamos uma distncia vetorial (por exemplo, L1, L2).
    - Essa distncia  o que ser usada para treinar o modelo com uma funo de perda contrastiva.
    - Dica: utilize a funo [F.pairwise_distance](https://docs.pytorch.org/docs/stable/generated/torch.nn.PairwiseDistance.html#torch.nn.PairwiseDistance).

 ```python
   distance = F.pairwise_distance(output1, output2)
   ```

````python
class BaseNetwork(nn.Module):
    def __init__(self):
        super(BaseNetwork, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2),
            
            nn.Conv2d(32, 64, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2),
            
            nn.Conv2d(64, 128, kernel_size=3),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2)
        )
        
        
        self.fc = nn.Sequential(
            nn.Linear(12800, 512), 
            nn.ReLU(inplace=True),
            nn.Linear(512, 128) 
        )

    def forward(self, x):
        x = self.conv(x)
        x = x.view(x.size(0), -1) 
        embedding = self.fc(x)
        return F.normalize(embedding, p=2, dim=1)

class SiameseNetwork(nn.Module):
    def __init__(self, base_network):
        super(SiameseNetwork, self).__init__()
        self.base_network = base_network

    def forward(self, x1, x2):
        # Gera os embeddings para as duas imagens
        emb1 = self.base_network(x1)
        emb2 = self.base_network(x2)
        
        # Calcula a Distncia Euclidiana
        distance = torch.sqrt(torch.sum((emb1 - emb2)**2, dim=1))
        return distance
````

## Funo de Perda

A contrastive loss penaliza pares semelhantes com distncia grande e pares diferentes com distncia pequena.



Implemente a funo contrastive_loss:

````python
def contrastive_loss(distance, label, margin=2.0):
    """
    Calcula a Contrastive Loss.
    label=1 para pares genunos, label=0 para pares impostores.
    """
    # Parte para pares da mesma pessoa: minimiza a distncia ao quadrado
    loss_genuine = label * torch.pow(distance, 2)
    
    # Parte para pares de pessoas diferentes: garante distncia mnima (margin)
    # torch.clamp(..., min=0.0) equivale ao max(0, margin - distance)
    loss_impostor = (1 - label) * torch.pow(torch.clamp(margin - distance, min=0.0), 2)
    
    # Retorna a mdia do batch
    return torch.mean(loss_genuine + loss_impostor)
````

## Treinando o Modelo

Se tudo estiver correto, o cdigo abaixo deve funcionar:

````python
from tqdm import tqdm

def train_model(model, train_loader, val_loader, optimizer, num_epochs, device):
    model.train()  # Coloca o modelo em modo de treino

    # Inicializa o dicionrio para armazenar as perdas
    history = {
        'train_losses': [],
        'val_losses': []
    }

    for epoch in range(num_epochs):
        running_loss = 0.0

        # Loop de treinamento com barra de progresso
        for i, (img1, img2, label) in enumerate(tqdm(train_loader, desc=f'Treino - poca {epoch + 1}/{num_epochs}', unit='batch')):
            img1, img2, label = img1.to(device), img2.to(device), label.to(device)

            # Forward pass 
            distance = model(img1, img2)

            # Calcular a perda
            loss = contrastive_loss(distance, label, margin = 1)

            # Backward pass e otimizao
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # Clculo da perda mdia para a poca
        avg_train_loss = running_loss / len(train_loader)
        history['train_losses'].append(avg_train_loss)  # Adiciona a perda de treinamento ao histrico
        print(f'Epoch [{epoch + 1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}')

        # Validao
        val_loss = validate_model(model, val_loader, device)
        history['val_losses'].append(val_loss)  # Adiciona a perda de validao ao histrico
        print(f'Epoch [{epoch + 1}/{num_epochs}], Validation Loss: {val_loss:.4f}')

    print('Treinamento completo')
    return history  # Retorna o dicionrio de histrico

def validate_model(model, val_loader, device):
    model.eval()  # Coloca o modelo em modo de avaliao
    running_loss = 0.0

    with torch.no_grad():  # No calcula os gradientes durante a validao
        for img1, img2, label in val_loader:
            img1, img2, label = img1.to(device), img2.to(device), label.to(device)
            output = model(img1, img2)
            loss = contrastive_loss(label, output)
            running_loss += loss.item()

    return running_loss / len(val_loader)  # Retorna a perda mdia de validao
````

````python
base_network = BaseNetwork().to(device)
model = SiameseNetwork(base_network).to(device)
optimizer = torch.optim.Adam(params= model.parameters(), lr=0.00001)
history = train_model(model, train_loader, validation_loader, optimizer, num_epochs=10, device=device)
````

````python
plt.plot(history['train_losses'], label='Training Loss')
plt.plot(history['val_losses'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.legend()
plt.show()
````

## Testando o Modelo

Para avaliar o modelo calculamos a loss no conjunto de teste e plotamos algumas amostras e suas distncias no espao latente:

````python
def test_model(model, test_loader, device):
    model.eval()  # Coloca o modelo em modo de avaliao (desativa dropout, etc.)
    running_test_loss = 0.0

    with torch.no_grad():  # Desativa o clculo de gradientes para economizar memria
        for img1, img2, label in test_loader:
            img1, img2, label = img1.to(device), img2.to(device), label.to(device)

            # Forward pass
            distance = model(img1, img2)

            # Calcular a perda
            loss = contrastive_loss(distance, label)
            running_test_loss += loss.item()

    # Clculo da perda mdia no conjunto de teste
    avg_test_loss = running_test_loss / len(test_loader)
    print(f'Test Loss: {avg_test_loss:.4f}')

    return avg_test_loss


def plot_test_images(model, test_loader, device, num_pairs=4):
    model.eval()  # Coloca o modelo em modo de avaliao
    fig, axes = plt.subplots(num_pairs, 2, figsize=(5, num_pairs * 2.5))  # Configurar o plot

    with torch.no_grad():  # Desativa gradientes
        count = 0  # Contador para nmero de pares mostrados
        for img1, img2, label in test_loader:
            if count >= num_pairs:  # Limitar a quantidade de pares
                break

            img1, img2, label = img1.to(device), img2.to(device), label.to(device)

            # Forward pass para obter a distncia para o batch inteiro
            distances = model(img1, img2)

            # Para cada par no batch
            for j in range(img1.size(0)):
                if count >= num_pairs:
                    break  # Interrompe se j mostramos o nmero necessrio de pares

                # Converter img1 e img2 para o formato correto (removendo a dimenso do batch)
                img1_np = img1[j].cpu().numpy()  # Selecionar a j-sima imagem no batch
                img2_np = img2[j].cpu().numpy()

                # Plotar as imagens
                axes[count, 0].imshow(img1_np.squeeze(), cmap='gray')
                axes[count, 1].imshow(img2_np.squeeze(), cmap='gray')



                axes[count, 0].set_title(f"Dissimilaridade: {distances[j].item():.2f} ")

                axes[count, 0].axis('off')
                axes[count, 1].axis('off')

                count += 1  # Incrementar contador

    plt.tight_layout()
    plt.show()

# Avaliando o modelo no conjunto de teste
test_loss = test_model(model, test_loader, device)
plot_test_images(model, test_loader, device)
````
