# **Pré-processamento de dados**

Primeiro, vamos manter o controle de algumas dimensões dos dados:
- ``img_rows`` que deve ser 28
- ``img_cols``, que deve ser 28
- ``input_shape``, que é 28 x 28 x 1

````python

img_rows = x_train[0].shape[0]
img_cols = x_train[0].shape[1]

input_shape = (img_rows, img_cols, 1)
````

````python
img_rows = x_train[0].shape[0] #Pegando a dimensão 28
img_cols = x_train[0].shape[1] #Pegando a dimensão 28

# Obtendo nossos dados na “forma” correta necessária para o Keras
# Precisamos adicionar uma quarta dimensão aos nossos dados, alterando assim nossa forma de imagem original de (60000,28,28) para (60000,28,28,1)

x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)

input_shape = (img_rows, img_cols, 1) #dimensões de uma unica imagem


x_train = x_train.astype('float32') #uint8 originally
x_test = x_test.astype('float32')

# Normalizar os valores de cada pixel para o intervalo de 0 a 1
x_train /= 255.0
x_test /= 255.0

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
````

### **One Hot Encode our Labels**
**Exemplo de One Hot Encoding**

![Image of one hot ncoding](https://raw.githubusercontent.com/rajeevratan84/DeepLearningCV/master/hotoneencode.JPG)
Além disso, mantenha as classes de números armazenadas como uma variável, ```num_classess```

````python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Contando o numero de colunas após aplicação do One Hot Encoder
print ("Número de classes: " + str(y_test.shape[1]))

num_classes = y_test.shape[1]
num_pixels = x_train.shape[1] * x_train.shape[2]
````

**Criando nosso modelo**

Esse é o mesmo CNN que usamos anteriormente para o projeto de classificação MNIST.

**Adicionando camadas de conv**

``Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape)``

Nosso **Conv2D()** cria o filtro com os seguintes argumentos:
- Número de filtros, usamos 32
- kernel_size, usamos um filtro 3x3, portanto, ele é definido como uma tupla ``(3,3)``
- ativação, em que especificamos ``'relu'``
- input_shape, que obtivemos e armazenamos em uma variável acima, é, em nosso exemplo, uma imagem em escala de cinza de 28 x 28. Portanto, nossa forma é ``(28,28,1)``

**Adicionando camadas MaxPool**

Novamente, usamos ``model.add()`` e especificamos ``MaxPooling2D(pool_size=(2,2))``.

Usamos o argumento de entrada pool_size para definir o tamanho da nossa janela. Podemos especificar stride e padding da seguinte forma:

``pool_size=(2, 2), strides=None, padding='valid'``

Entretanto, observe que o stride padrão é usado como o tamanho da janela de pooling (2 no nosso caso). 

Usar ``padding ='valid'`` significa que não usamos padding.

**Adicionando Flatten**

Usando model.add(Flatten()), estamos simplesmente achatando a saída do nosso último nó. O que equivale a 12 x 12 * 64 * 1 = 9216.

**Adicionando camadas densas ou totalmente conectadas**

``model.add(Dense(128, activation='relu'))``

Usamos ``model.add()`` mais uma vez e especificamos o número de nós aos quais a camada anterior se conectará. Também especificamos a função de ativação do ReLU aqui.





````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import backend as K
from tensorflow.keras.optimizers import SGD 


model = Sequential()

model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))

model.add(Conv2D(64, (3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(num_classes, activation='softmax'))

loss = tf.keras.losses.get("categorical_crossentropy")

def loss_wrapped_with_function(*args, **kwargs):
    return loss(*args, **kwargs)

model.compile(loss = loss,
              optimizer = SGD(0.001, momentum=0.9),
              metrics = ['accuracy'])

print(model.summary())
````

# **Treinamento do nosso modelo**

````python
x_test.shape
````

````python

batch_size = 32
epochs = 15

history = model.fit(x_train,
                    y_train,
                    batch_size = batch_size,
                    epochs = epochs,
                    verbose = 1,
                    validation_data = (x_test, y_test))

# Obtemos nossa pontuação de acurácia usando a função evalute
# A pontuação contém dois valores, nossa perda de teste e a precisão
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
````

# **Exemplo de aumento de dados** 

Usamos ``generators`` porque não podemos carregar todo o conjunto de dados na memória de nossos sistemas. Portanto, usamos esse recurso para criar um iterador, de modo que possamos acessar lotes de nossos dados para aumento de dados ou pré-processamento e propagação direta em nossa rede.

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Recarregar nossos dados
(x_train, y_train), (x_test, y_test)  = fashion_mnist.load_data()

# Reformule nossos dados para que fiquem no formato [número de amostras, largura, altura, profundidade_de_cor]
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

# Alterar o tipo de dados para float32
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# Definir gerador o aumento de dados
data_aug_datagen = ImageDataGenerator(rotation_range=30,
                                      width_shift_range=0.1,
                                      height_shift_range=0.1,
                                      shear_range=0.2,
                                      zoom_range=0.1,
                                      horizontal_flip=True,
                                      fill_mode='nearest')

# Create our iterator
aug_iter = data_aug_datagen.flow(x_train[0].reshape(1,28,28,1), batch_size=1)
````

````python
import cv2


def showAugmentations(augmentations = 6):
    fig = plt.figure()
    for i in range(augmentations):
        a = fig.add_subplot(1,augmentations,i+1)
        img = next(aug_iter)[0].astype('uint8')
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')

showAugmentations(6)
````
